## Image pre-processing
`shapefile` and `.tiff` preprocessed in QGIS, downloaded from OpenStreetMap

```python
# Importing libraries
import fiona
import rasterio as rio
import numpy as np
import matplotlib.pyplot as plt
from rasterio.plot import show
import rasterio.mask

%matplotlib inline
```

```python
# Use fiona to read in shapefile and vectorize it
with fiona.open('../data/dennery.shp', 'r') as shapefile:
    shapes = [feature['geometry'] for feature in shapefile]
```

```python
# Use rasterio to open source image and apply the vector mask
with rasterio.open("../data/st_lucia.tif") as src:
    out_image, out_transform = rasterio.mask.mask(src, shapes, crop=True)
    out_meta = src.meta
```

```python
# Checking resulting meta values
out_meta
```

```python
# Plotting resulting masked raster
fig, ax = plt.subplots(figsize = (15,15))
show(out_image);
```

```python
# Updating metadata for output
out_meta.update({"driver": "GTiff",
                 "height": out_image.shape[1],
                 "width": out_image.shape[2],
                 "transform": out_transform})

# Saving the resulting mask as a .tiff file for use in QGIS
with rasterio.open("../data/st_lucia_masked.tif", "w", **out_meta) as dest:
    dest.write(out_image)
```

## Model
Model instantiation and training using the resulting QGIS image and label tiles

```python
# Importing tensorflow and Keras
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet50 import ResNet50
```

```python
# Directories of image and label chips (file structure must match specifications in ResNet documentation)
image_directory = '../../data/224_chips/images/'
label_directory = '../../data/224_chips/labels/'

# Image size
img_rows, img_cols = 224, 224

# Number of channels (RGB)
img_channels = 3

# One class - building
nb_classes = 1

# Instantiate ImageDataGenerator
gen = ImageDataGenerator(rescale=1./255)
```

```python
# Function to create generator
def create_gen(generator, directory, gen_type):
    # Parameters for 'labels' vs images:
    if gen_type == 'label':
        return generator.flow_from_directory(directory,
                                             class_mode = None,
                                             shuffle = False,
                                             color_mode = 'grayscale',
                                             target_size=(img_rows, img_cols),
                                             batch_size = 3
                                             )
    else:
        return generator.flow_from_directory(directory,
                                             class_mode = None,
                                             shuffle = False,
                                             target_size=(img_rows, img_cols),
                                             batch_size = 3
                                             )

# Function to combine generators for passing into the model
def final_gen(image_gen, label_gen):
    for data, labels in zip(image_gen, label_gen):
        # divide labels by 255 to make them like masks i.e. 0 and 1
        labels /= 255.
        # remove the last axis, i.e. (batch_size, n_rows, n_cols, 1) --> (batch_size, n_rows, n_cols)
        labels = np.squeeze(labels, axis=-1)

        yield data, labels
```

```python
# Image generators
train_image_gen = create_gen(gen, image_directory + 'train/', 'image')
test_image_gen = create_gen(gen, image_directory + 'validation/', 'image')

# Label generators
train_label_gen = create_gen(gen, label_directory + 'train/', 'label')
test_label_gen = create_gen(gen, label_directory + 'validation/', 'label')
```

```python
# Final train and test generators
train_gen = final_gen(train_image_gen, train_label_gen)
test_gen = final_gen(test_image_gen, test_label_gen)
```

```python
# Instantiate model (ResNet50) with default parameters
model = ResNet50() 
```

```python
# Compile with selected loss function, optimizer and metrics
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Fit model with the created generator
model.fit_generator(
        train_gen,
        steps_per_epoch = 500,
        epochs=3)
#         Commented out for testing the model
#         validation_data=validation_generator,
#         validation_steps=250)
```
